```python
import time
import sys, os

from pennylane import numpy as np
import pandas as pd
import pennylane as qml

from sklearn.decomposition import PCA, NMF
from sklearn.preprocessing import MinMaxScaler, StandardScaler

import matplotlib.pyplot as plt
```

```python
data_file_path = 'datasets/unsw-nb15/'
```

```python
df = pd.read_csv(f'{data_file_path}Payload_data_UNSW.csv')
```

```python
labels = df.pop('label')
protocols = df.pop('protocol')
t_deltas = df.pop('t_delta')

scaler = MinMaxScaler()
X = scaler.fit_transform(df)        

nmf = NMF(n_components=50, init='nndsvda', max_iter=500, random_state=42)
df_pca   = nmf.fit_transform(X)        
```

```python
df_pca = pd.DataFrame(df_pca)
```

```python
df_pca['protocol'] = protocols
df_pca['t_delta'] = t_deltas
df_pca['label'] = labels
```

```python
def one_hot_encoding(df, column):
    df = pd.concat([df, pd.get_dummies(df[column], prefix=column)], axis=1)
    df.drop(column, axis=1, inplace=True, errors='ignore')
    return df
```
